```python
import pandas as pd

def standardize_data(input_file, output_file):
    print(f"Reading data from {input_file}...")
    df = pd.read_csv(input_file)
    
    print("\nColumn names:", df.columns.tolist())

    df['Bird Species'] = df['Bird Species'].str.strip()
    
    print(f"\nSaving standardized data to {output_file}...")
    df.to_csv(output_file, index=False)
    
    print(f"\nStandardization complete. Data saved to {output_file}")
    
    return df
```

```python
standardize_data('raw_data/2016_CBCM_birds_willowbrook.csv', 'standard_data/2016_CBCM_birds_willowbrook.csv')
```
